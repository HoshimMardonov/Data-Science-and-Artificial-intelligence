```python
import random 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
from sklearn.datasets import make_blobs 
%matplotlib inline
```

<h1>Clustering</h1>


In this exercise, we generate random points and cluster these points. Objective: To see how the clustering algorithm works:

We use the make_blobs function to create a cluster of random blobs. This function accepts the following parameters:

**n_samples** is the number of points

**centers** - coordinates of the cluster center (centroid) -cluster_std - standard deviation from the center

The **make_blobs** function returns the x and y coordinates of the blobs.

```python
np.random.seed(0)
centroids = [[3,3], [-3,-2], [2,-3],[0,0]]
X, y = make_blobs(n_samples=5000, centers=centroids, cluster_std=0.8)
plt.scatter(X[:,0], X[:,1], marker='.')
plt.show()
```

<h2>k-means algorithm</h2>

The KMeans function accepts the following parameters:

**init** - centroid selection method (k-means++ or random)

**n_clusters** is the number of clusters

**n_init** - number of times to run the algorithm (by running it repeatedly with different centroids, the best centroids are selected to start building the model)

```python
k_means = KMeans(init = "random", n_clusters = 4, n_init = 15)
```

```python
k_means.fit(X)
```

Points were clustered. To see the cluster number, we refer to the .labels_ parameter.

```python
k_means.labels_
```

To see the cluster centers, we refer to the .cluster_centers_ parameter.

```python
k_means.cluster_centers_
```

We compare the cluster centers found by the algorithm with the centers we gave earlier:

```python
print(centroids)
print(np.round(k_means.cluster_centers_))
```

```python
fig = plt.figure(figsize=(8, 8))
# Har bir klaster uchun alohida rang
colors = plt.cm.Spectral(np.linspace(0, 1, len(set(k_means.labels_))))
ax = fig.add_subplot(1, 1, 1)

for k, col in zip(range(len([[3,3], [-3, -2], [2, -3], [0, 0]])), colors):

    # Create a list of all data points, where the data points that are 
    # in the cluster (ex. cluster 0) are labeled as true, else they are
    # labeled as false.
    my_members = (k_means.labels_ == k)
    
    # Define the centroid, or cluster center.
    cluster_center = k_means.cluster_centers_[k]
    
    # Plots the datapoints with color col.
    ax.plot(X[my_members, 0], X[my_members, 1], 'w', markerfacecolor=col, marker='.')
    
    # Plots the centroids with specified color, but with a darker outline
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,  markeredgecolor='k', markersize=6)

# Title of the plot
ax.set_title('KMeans')

# Remove x-axis ticks
ax.set_xticks(())

# Remove y-axis ticks
ax.set_yticks(())

# Show the plot
plt.show()
```
